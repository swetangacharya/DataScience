**Latent Dirichlet Allocation**  

LDA is a generative probabilistic model


Each document can be described by a distribution of topics and each topic can be described by a distribution of words. Now, you have many documents that are talking similar topics. and a document can have one or more topics. So, if you have average 500 words and 1000 documents, then you need 500x1000 threads to connect to them. when you see that some documents are connected to same set of words. You know they discuss the same topic. Then you can read one of those documents and know what all these documents talk about.   
To solve this, we need another variable that is latent(hidden) and contains list of words that describe similar topic and connect these topics to a document. For example if certain set of words that has high probabilities in a given topic and they are presented in document then we can say that the document represent that topic. A document can have one or more such topic.   
So, topic like Animal can have probability set like this.\{0.3Cats,0.4Dogs,0.2Loyal, 0.1Evil,...\}

**$\alpha$** :A hyperparameter for the Dirichlet prior on the per-document topic distribution ($\theta$).   if $\alpha$ is high then Documents are assumed to contain a mix of most topics (more topic diversity). If it is low then Documents are likely to contain only a few or one topic (less topic diversity).  
**$\theta$ - Document-Topic Distribution:**
it is a random distribution (often a matrix) representing the topic distribution for a specific document.
function: $\theta_{d,k}$ gives probability that document d contains topic k. it is sampled from Dirichlet distribution defined by $\alpha$

$\beta$: Governs distribution of words in topics (prior).A hyperparameter for the Dirichlet prior on the per-topic word distribution ($φ$ or $\beta$), if $\beta$ is high then Topics are composed of many words from the vocabulary. If it is low then Topics are composed of only a few specific words (more specific topics)  
$\eta$ is the parameter for the Dirichlet prior that dictates how topics are generated over words. It determines the "smoothness" of word distributions within topics.   

Low $\eta$ , Each topic will contain a small, specific set of words. This makes topics more distinct and sparse.   
High $\eta$, Each topic will contain a mixture of most words in the vocabulary, making topics more similar to each other.     

$\theta$ is a random matrix $\theta(i,j)$ represents probability of ith document containing words belonging to jth topic. (MxK)(M=number of documents and K is number of topics).   

similarly,$\beta(i,j)$ represents probability of ith topic containing jth word. both $\theta$ and $\beta$ are Dirichlet distribution.
